## Combining english datasets to one big dataset

```python
import pandas as pd
```

```python
weather = pd.read_csv("../Data/english_datasets/weather_sentiment_confidence.csv")
```

```python
weather
```

```python
airlines = pd.read_csv("../Data/english_datasets/Airline-Sentiment-2-w-AA.csv", encoding='latin-1')
```

```python
airlines.head(20)
```

```python
self_drive = pd.read_csv("../Data/english_datasets/Twitter-sentiment-self-drive-DFE.csv", encoding='latin-1')
```

```python
self_drive
```

```python
apple = pd.read_csv("../Data/english_datasets/Apple-Twitter-Sentiment-DFE.csv", encoding='latin-1')
```

```python
apple
```

```python
sentences = pd.read_csv("../Data/english_datasets/primary-plutchik-wheel-DFE.csv", encoding='latin-1')
```

```python
sentences
```

```python
weather = weather[['text','Confidence']]
```

```python
weather.columns = ['text','conf']
weather.hist(bins=100)
```

```python
airlines = airlines[['text','airline_sentiment:confidence']]
```

```python
airlines.columns = ['text','conf']
airlines.hist(bins=100)
```

```python
self_drive = self_drive[["text","sentiment:confidence"]]
```

```python
self_drive.columns = ['text','conf']
self_drive.hist(bins=100)
```

```python
apple = apple[["text","sentiment:confidence"]]
```

```python
apple.columns = ['text','conf']
apple.hist(bins=100)
```

```python
sentences = sentences[["sentence","emotion:confidence"]]
```

```python
sentences.columns = ['text','conf']
sentences.hist(bins=100)
```

```python
bigdf = pd.concat([weather, self_drive, apple, sentences], ignore_index=True)
```

```python
bigdf
```

```python
bigdf.to_pickle("../Data/english_datasets/english_sentiment_with_confidence.pkl")
```
